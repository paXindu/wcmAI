```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Load dataset
df = pd.read_csv('new.csv')
```

```python
df.head()
```

```python
location = pd.get_dummies(df.Location)
merged= pd.concat([df,location],axis='columns')
merged
```

```python
x = merged.drop(['fillingrate','Location','Location 1'], axis=1)
```

```python
x
```

```python
y = df['fillingrate']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
pred
```

```python
x_test
```

```python
y_test
```

```python
pred = model.predict(x_test)

# Compute mean squared error
mse = mean_squared_error(y_test, pred)

# Print mean squared error
print("Mean Squared Error:", mse)
```

```python
# Ask user for temperature, rainfall, and location input
print("Enter temperature, rainfall, and location (separated by commas):")
user_input = input().split(",")
user_input = [float(x.strip()) if i < 2 else x.strip() for i, x in enumerate(user_input)]

# Get one-hot encoding for user's location input
user_location = pd.get_dummies([user_input[2]])[merged.columns[2:]]
user_location = user_location.reindex(columns=x.columns, fill_value=0)

# Create input array for model prediction
user_x = np.hstack([user_input[:2], user_location])

# Use model to predict filling rate
user_filling_rate = model.predict(user_x.reshape(1, -1))

# Print predicted filling rate
print("Predicted filling rate:", user_filling_rate[0])
```
