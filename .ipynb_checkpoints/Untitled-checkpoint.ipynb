```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("colombodata.csv")

# Select the features
X = df[['Temperature', 'Rainfall']]
y = df['filling rate']

# Create a Linear Regression model
model = LinearRegression()

# Train the model on the data
model.fit(X, y)

# Make predictions using the model
y_pred = model.predict(X)

# Create a scatter plot of the data
plt.scatter(X['Temperature'], y, label='Temperature')
plt.scatter(X['Rainfall'], y, label='Rainfall')

# Add the regression line
plt.plot(X['Temperature'], y_pred, color='green')
plt.plot(X['Rainfall'], y_pred, color='blue')

# Add labels and legend
plt.xlabel('Temperature and Rainfall')
plt.ylabel('Filling rate')
plt.legend()

# Show the plot
plt.show()
```

```python
model.fit(X, y)

# Make a complex prediction
new_data = [[27.5, 150], [22, 200]]
prediction = model.predict(new_data)

# Print the prediction
print(prediction)
```
